```python
%matplotlib inline
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
impo

dataset_path = 'G:\TACO_Data\data'
anns_file_path = dataset_path + '/' + 'annotations.json'

# Read annotations
with open(anns_file_path, 'r') as f:
    dataset = json.loads(f.read())

categories = dataset['categories']
anns = dataset['annotations']
imgs = dataset['images']
nr_cats = len(categories)
nr_annotations = len(anns)
nr_images = len(imgs)

# Load categories and super categories
cat_names = []
super_cat_names = []
super_cat_ids = {}
super_cat_last_name = ''
nr_super_cats = 0
for cat_it in categories:
    cat_names.append(cat_it['name'])
    super_cat_name = cat_it['supercategory']
    # Adding new supercat
    if super_cat_name != super_cat_last_name:
        super_cat_names.append(super_cat_name)
        super_cat_ids[super_cat_name] = nr_super_cats
        super_cat_last_name = super_cat_name
        nr_super_cats += 1

print('Number of super categories:', nr_super_cats)
print('Number of categories:', nr_cats)
print('Number of annotations:', nr_annotations)
print('Number of images:', nr_images)
```

```python
for i in range(1,16): 
    for filename in os.listdir('G:\TACO_Data\data'+'/'+'batch_'+ str(i)):
        
        if filename.startswith("0") or filename.startswith("I") :  # 
            
            image_path = os.path.join('G:\TACO_Data\data'+'/'+'batch_'+ str(i), filename)
            
            jpg_file = os.path.join('batch_'+ str(i)+'/'+os.path.splitext(filename)[0] + '.jpg')
            JPG_file = os.path.join('batch_'+ str(i)+'/'+os.path.splitext(filename)[0] + '.JPG')
            
            txt_file = os.path.join('batch_'+ str(i)+'_'+os.path.splitext(filename)[0] + '.txt')
            
        with open(anns_file_path, 'r') as f:
            data = json.load(f)
        
        for fn in data['images']:
            if (fn['file_name'] == jpg_file) or (fn['file_name'] == JPG_file) :
                # txt
                with open('G:\\taco-yolo8\\label\\' + txt_file, 'w') as f:
                    for ann in anns:
                        if ann['image_id'] == fn['id']:
                            class_id = ann['category_id']
                            class_name= label_map[class_id]
                            x, y, w, h = ann['bbox']
                            img_width = fn['width']
                            img_height = fn['height']
                            center_x = x / img_width
                            center_y = y / img_height
                            width = w / img_width
                            height = h / img_height
                            f.write(f"{class_id} {center_x:.6f} {center_y:.6f} {width:.6f} {height:.6f}\n")
```

```python
import os
import shutil

source_folder = "G:\\TACO_Data\\data\\batch_2"
target_folder = "G:\\taco-yolo8\\data\\image"
prefix = "batch_2_"

for j in range(1,16):
    source_folder = "G:\\TACO_Data\\data\\batch_"+str(j)
    target_folder = "G:\\taco-yolo8\\data\\image"
    prefix = 'batch_'+str(j)+'_'

    image_files = [f for f in os.listdir(source_folder) if f.startswith('0') or f.startswith('I')]
    # 
    for i, file_name in enumerate(image_files):
        # 
        new_file_name = prefix + file_name
        # 
        shutil.copyfile(os.path.join(source_folder, file_name), os.path.join(target_folder, new_file_name))
```
